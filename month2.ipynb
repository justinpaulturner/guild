```python
import pandas as pd
from datetime import datetime as dt
import datetime
```

```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/spreadsheets.readonly']

# The ID and range of a sample spreadsheet.
SPREADSHEET_ID = '1fdfhA4n_VOJOzExVfB5p19JmLJzQGFq1n81cUj6wjcQ'
# month 2: 3/15 - 4/14
# month 3: 4/15 - 5/15
SAMPLE_RANGE_NAME = 'Form Responses 1!A252:F600'


def get_spreadsheet():
    """Shows basic usage of the Sheets API.
    Prints values from a sample spreadsheet.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    try:
        service = build('sheets', 'v4', credentials=creds)

        # Call the Sheets API
        sheet = service.spreadsheets()
        result = sheet.values().get(spreadsheetId=SPREADSHEET_ID, range=SAMPLE_RANGE_NAME).execute()
        values = result.get('values', [])
    except HttpError as err:
        print(err)
        
    return values
```

```python
dt_str = '3/18/2023 9:13:09'
def get_dt(dt_str):
    month, day, year = dt_str.split(" ")[0].split("/")
    month, day = month.zfill(2), day.zfill(2)
    hour, minute, second = dt_str.split(" ")[1].split(":")
    hour, minute, second = hour.zfill(2), minute.zfill(2), second.zfill(2)
    cleaned_dt_str = f"{month}/{day}/{year} {hour}:{minute}:{second}"
    return dt.strptime(cleaned_dt_str, "%m/%d/%Y %H:%M:%S")
```

```python
def get_range_list(start, end):
    day_list = pd.period_range(start=start, end=end, freq='D')
    day_list = [day.strftime("%m/%d/%Y") for day in day_list]
    return day_list
```

```python
group_data = {name: {
    "litany_misses": set(), 
    "habit_misses": set(),
    "no_snacking_misses": set(),
    "litany_successes": set(),
    "habit_successes": set(),
    "no_snacking_successes": set()
} for name in names}
group_data

for row in corrected_rows:
    date, name, litany, habit, yesterday, no_snacking = row
    dt_obj = get_dt(date)
    if yesterday and dt_obj.hour < 12:
        corrected_date = (dt_obj - datetime.timedelta(1))
        dt_obj = corrected_date
    if yesterday and dt_obj.hour > 12:
        litany, habit, no_snacking = "", "", ""
    month, day, year = str(dt_obj.month).zfill(2), str(dt_obj.day).zfill(2), dt_obj.year
    if litany == "":
        group_data[name]["litany_misses"].add(f"{month}/{day}/{year}")
    else:
        group_data[name]["litany_successes"].add(f"{month}/{day}/{year}")
    if habit == "":
        group_data[name]["habit_misses"].add(f"{month}/{day}/{year}")
    else:
        group_data[name]["habit_successes"].add(f"{month}/{day}/{year}")
    if no_snacking == "":
        group_data[name]["no_snacking_misses"].add(f"{month}/{day}/{year}")
    else:
        group_data[name]["no_snacking_successes"].add(f"{month}/{day}/{year}")

for name, info in group_data.items():
    info["litany_success_count"] = len(info["litany_successes"])
    info["habit_success_count"] = len(info["habit_successes"])
    info["no_snacking_success_count"] = len(info["no_snacking_successes"])
```

```python
date_range = get_range_list(corrected_rows[0][0], corrected_rows[-1][0])
days_count = len(date_range)
```

```python
# update group_data for missed days
for date in date_range:
    for name, info in group_data.items():
        for habit in habits:
            if date not in info[f"{habit}_successes"]:
                info[f"{habit}_misses"].add(date)
```

```python
total_litany_miss_count = 0
total_habit_miss_count = 0
total_no_snacking_miss_count = 0
total_litany = 0
total_habit = 0
total_no_snacking = 0
for name, info in group_data.items():
    if info["litany_misses"]:
        litany_misses = len(info["litany_misses"])
        total_litany_miss_count += litany_misses
        litany_misses_percent = litany_misses / days_count
        info["litany_miss_percentage"] = f"{round(litany_misses_percent, 2)}"
    else:
        info["litany_miss_percentage"] = "0"
    total_litany += info["litany_success_count"]
    if info["habit_misses"]:
        habit_misses = len(info["habit_misses"])
        total_habit_miss_count += habit_misses
        habit_misses_percent = habit_misses / days_count
        info["habit_miss_percentage"] = f"{round(habit_misses_percent, 2)}"
    else:
        info["habit_miss_percentage"] = "0"
    total_habit += info["habit_success_count"]
    if info["no_snacking_misses"]:
        no_snacking_misses = len(info["no_snacking_misses"])
        total_no_snacking_miss_count += no_snacking_misses
        no_snacking_misses_percent = no_snacking_misses / days_count
        info["no_snacking_miss_percentage"] = f"{round(no_snacking_misses_percent, 2)}"
    else:
        info["no_snacking_miss_percentage"] = "0"
    total_no_snacking += info["no_snacking_success_count"]

average_litany_miss = (total_litany_miss_count / days_count) / len(names)
average_habit_miss = (total_habit_miss_count / days_count) / len(names)
average_no_snacking_miss = (total_no_snacking_miss_count / days_count) / len(names)
group_info = {"Group Info": {
    'litany_miss_percentage': round(average_litany_miss, 2), 
    'habit_miss_percentage': round(average_habit_miss, 2),
    'no_snacking_miss_percentage': round(average_no_snacking_miss, 2),
    "total_litany_days": total_litany,
    "total_habit_days": total_habit,
    "total_no_snacking_days": total_no_snacking
}}
```
